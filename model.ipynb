# Data Shuffling + Separation
## run only ONCE!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
# separate 70 training, 30 validation
data = pd.read_csv('data/master.csv', sep=',')
data_np = data.values

for i in range(5):
    np.random.shuffle(data_np)

ratio=int(data_np.shape[0]*0.7)

# train
train_data = data_np[0:ratio]
df_train = pd.DataFrame(train_data)
df_train.to_csv('data/train.csv',sep=',')

# test
test_data = data_np[ratio:]
df_test = pd.DataFrame(test_data)
df_test.to_csv('data/test.csv',sep=',')
```

# Data Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
# import training data
train = pd.read_csv('data/train.csv',sep=',')

# dropped first column = index
train = train.drop(train.columns[0], axis=1)

# dropped HDI column due to incomplete data
train = train.drop(train.columns[8], axis=1)

display(train)
```

# Data Visualization

```python
# srate vs. year
sry = pd.DataFrame(train.iloc[:,1])
srate = train.iloc[:,6]
sry = sry.join(srate)


sry = sry.groupby('1').mean()
display(sry)

plt.figure(figsize=(40,30))
plt.plot(sry.index,sry['6'])
plt.xlabel('year',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# s vs. year
sy = pd.DataFrame(train.iloc[:,1])
s = train.iloc[:,4]
sy = sy.join(s)


sy = sy.groupby('1').mean()
display(sy)

plt.figure(figsize=(40,30))
plt.plot(sy.index,sy['4'])
plt.xlabel('year',fontsize=20)
plt.ylabel('mean suicide #',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# srate vs. country
src = pd.DataFrame(train.iloc[:,0])
srate = train.iloc[:,6]
src = src.join(srate)


src = src.groupby('0').mean()
src = src.sort_values(by='6')
display(src)

plt.figure(figsize=(40,30))
plt.bar(src.index,src['6'])
plt.xlabel('country',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(rotation=90,fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# srate vs. age
sra = pd.DataFrame(train.iloc[:,3])
srate = train.iloc[:,6]
sra = sra.join(srate)

sra = sra.groupby('3').mean()
sra = sra.sort_values(by='6')
display(sra)

plt.figure(figsize=(40,30))
plt.bar(sra.index,sra['6'])
plt.xlabel('age group',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# srate vs. sex
srx = pd.DataFrame(train.iloc[:,2])
srate = train.iloc[:,6]
srx = srx.join(srate)

srx = srx.groupby('2').mean()
display(srx)

plt.figure(figsize=(40,30))
plt.bar(srx.index,srx['6'])
plt.xlabel('sex',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# srate vs. gdp/capita
srg = pd.DataFrame(train.iloc[:,9])
srate = train.iloc[:,6]
srg = srg.join(srate)
display(srg)

plt.figure(figsize=(40,30))
plt.scatter(srg['10'],srg['6'])
plt.xlabel('gdp/capita',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
# srate vs. generation
srgen = pd.DataFrame(train.iloc[:,10])
srate = train.iloc[:,6]
srgen = srgen.join(srate)

srgen = srgen.groupby('11').mean()
display(srgen)

plt.figure(figsize=(40,30))
plt.bar(srgen.index,srgen['6'])
plt.xlabel('generation',fontsize=20)
plt.ylabel('mean suicide #/100 000',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
